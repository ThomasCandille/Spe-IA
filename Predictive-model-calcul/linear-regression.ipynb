```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 20 random values for square meters and corresponding prices
x = np.random.randint(50, 300, 20)  # Random square meters between 50-300
y = x * np.random.uniform(800, 1200, 20)# Price based on sq meters

coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) # poly1d_fn is now a function which takes in x and returns an estimate for y

plt.plot(x,y, 'yo', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker
plt.xlim(0, 320)
plt.ylim(0, 400000)

value_to_predict = np.random.randint(50, 300)
predicted_price = poly1d_fn(value_to_predict)
plt.plot(value_to_predict, predicted_price, 'ro') #red circle marker
print(f"Predicted price for {value_to_predict} square meters: ${predicted_price:.2f}")
```
