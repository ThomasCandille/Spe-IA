```python
# Import necessary libraries
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the image
image = Image.open('chickentree.jpg')

# Display image information
print("Image loaded successfully!")
print(f"Image mode: {image.mode}")
print(f"Image format: {image.format}")
```

```python
# Display the image
plt.figure(figsize=(10, 8))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# 1. Horizontal flip
image_flipped = image.transpose(Image.FLIP_LEFT_RIGHT)

plt.figure(figsize=(10, 8))
plt.imshow(image_flipped)
plt.axis('off')
plt.show()
```

```python
# 2. Tilt the image
angle = np.random.randint(-30, 31)  # Random angle between -30 and 30 degrees
image_tilted = image_flipped.rotate(angle, expand=True, fillcolor='white')

plt.figure(figsize=(10, 8))
plt.imshow(image_tilted)
plt.axis('off')
plt.show()
```

```python
# 3. Add Gaussian noise
# Convert image to numpy array
image_array = np.array(image_tilted, dtype=np.float32)

# Generate Gaussian noise
mean = 0
std = 25  # Standard deviation (adjust for more/less noise)
gaussian_noise = np.random.normal(mean, std, image_array.shape)

# Add noise to image
noisy_image = image_array + gaussian_noise

# Clip values to valid range [0, 255]
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

# Convert back to PIL Image
image_final = Image.fromarray(noisy_image)

plt.figure(figsize=(10, 8))
plt.imshow(image_final)
plt.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 8))

axes[0].imshow(image)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(image_final)
axes[1].set_title('Final Image with Noise')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
